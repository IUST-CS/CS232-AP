```cpp
#include <fstream>
#include <cstdlib>
#include <cctype>

// Class for amounts of money in U.S. currency.
class Money
{
public:
    friend Money operator+(const Money &amount1, const Money &amount2);
    friend Money operator -(const Money &amount1, const Money &amount2);
    friend Money operator -(const Money &amount);
    friend bool operator==(const Money &amount1, const Money &amount2);
    Money(long dollars, int cents);
    Money(long dollars);
    Money();
    double getValue() const;
    friend istream &operator>>(istream &ins, Money &amount);
    // Overloads the >> operator so it can be used to input values of type Money.
    // Notation for inputting negative amounts is as in -$100.00.
    // Precondition: If ins is a file input stream, then ins has already been
    // connected to a file.
    friend ostream &operator<<(ostream &outs, const Money &amount);
    // Overloads the < operator so it can be used to output values of type Money.
    // Precedes each output value of type Money with a dollar sign.
    // Precondition: If outs is a file output stream,
    // then outs has already been connected to a file.
private:
    long allCents;
};
int digitToInt(char c);
// Used in the definition of the overloaded input operator >>.
// Precondition: c is one of the digits '0' through '9'.
// Returns the integer for the digit; for example, digitToInt('3') returns 3.

{
    Money amount;

    }

    outStream.open("outfile.dat");

istream &operator>>(istream &ins, Money &amount)
{
    char oneChar, decimalPoint,
        digit1, digit2; // digits for the amount of cents
    long dollars;
    int cents;
    bool negative; // set to true if input is negative.
    ins >> oneChar;
    if (oneChar == '-')
    {
        negative = true;
        ins >> oneChar; // read '$'
    }
    else
        negative = false;
    // if input is legal, then oneChar == '$'
    ins >> dollars >> decimalPoint >> digit1 >> digit2;
    if (oneChar != '$' || decimalPoint != '.' || !isdigit(digit1) || !isdigit(digit2))
    {
        cout << "Error illegal form for money input\n";
        exit(1);
    }
    cents = digitToInt(digit1) * 10 + digitToInt(digit2);
    amount.allCents = dollars * 100 + cents;
    if (negative)
        amount.allCents = -amount.allCents;
    return ins;
}
int digitToInt(char c)
{
    return (static_cast<int>(c) - static_cast<int>('0'));
}

    if (amount.allCents < 0)
        outs << "- $" << dollars << '.';
    else
        outs << "$" << dollars << '.';

    if (cents < 10)
        outs << '0';
    outs << cents;

    return outs;
}
```

```cpp
    Money difference[5];
    for (i = 0; i < 5; i++)
        difference[i] = max - amount[i];
    cout << "The highest amount is " << max << endl;
    cout << "The amounts and their\n"
         << "differences from the largest are:\n";
    for (i = 0; i < 5; i++)
    {
        cout << amount[i] << " off by "
             << difference[i] << endl;
    }
    return 0;
}
```

```c++
void StringVar::operator= (const StringVar& rightSide)
{
    int newLength = strlen(rightSide.value);
     
    if (newLength > maxLength)
        newLength = maxLength;
    
    for (int i = 0; i < newLength; i++)
```

<h2 id="Information Hiding" style="color: #d42c06;">Problems with =</h2>
The definition of operator <code>=</code> has a problem
<ul style="margin-right:50px">
<li>Usually we want a copy of the right hand
argument regardless of its size.
<li>To do this, we need to delete the dynamic array in the left hand argument and allocate a new array large enough for the right hand side's dynamic array.

<h2 id="Information Hiding" style="color: #d42c06;">A New Problem With =</h2>
The new definition of operator <code>=</code> has a problem
<ul style="margin-right:50px">
<li>What happens if we happen to have the same object on each side of the assignment operator?
<ul>

```c++
    myString = myString;
```
</ul>

<li>This version of operator = first deletes the dynamic array in the left hand argument.
<li>Since the objects are the same object, there is no longer an array to copy from the right hand side!

<h2 id="Information Hiding" style="color: #d42c06;">A Better = Operator</h2>

```c++
    void StringVar::operator = (const StringVar& rightSide) 
    {
        int newLength = strlen(rightSide.value);
        if (newLength > maxLength) //delete value only if more space is needed
        { 
            delete [ ] value;
            maxLength = newLength;
            value = new char[maxLength + 1];
        }

        for (int I = 0; i< newLength; i++)
            value[i] = rightSide.value[i]; 
        
        value[newLength] = '\0';
    }
```
